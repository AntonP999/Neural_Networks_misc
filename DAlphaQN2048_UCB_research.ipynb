```python
def tree_search(root, state, net, number):
    # Поиск по дереву из состояния заданное число раз    
    for i in range(number):
        leaf = root.play_to_leaf() # Проход до нового листа (Select, Expand)        
        leaf.backup(leaf) # Backup
    return root
```

Выводы параметров корневого нода после каждого прохода до создания листа (128 итераций). На графиках представлены изменения доверительных интервалов для Q для каждого действия. Красным кружком обозначен softmax(log(Q*)), синим квадратом - верхняя граница интервала при отсутствии нейросети (метод UCB1), зеленым треугольником - верхняя граница при использовании нейросети. 

Позиция задана так, что оптимальное действие - налево (N[2]). Второе - направо (N[0]).

Первый вариант - необученная сеть.

```python
net = Alpha2048net(device=device)
net_optim = optim.Adam(net.parameters(), lr=1e-6, weight_decay=1e-5)
net.eval()

state = np.array([[0,0,0,0],[0,2,2,0],[4,0,0,8],[8,0,0,2]])
root = TreeNode(Game_Core_2048(), state, 0, True, net)
result = tree_search(root, state, net, 128)
```

По накопленным Q* видно, что MCTS на 128 итерациях находит лучшие ходы. P здесь служит модификатором доверительных интервалов. Предположительно, при хорошо обученной сети, она должна направлять поиск в потенциально выгодные направления. Свежеинициализированная сеть просто сжимает их на 0.25, при этом константа c_puct меняет величину сжатия, поощряя либо разведку, либо эксплуатацию.
